```python
class Student:
    def __init__(self, name, surname, age, city):
        self.name = name
        self.surname = surname
        self.age = age
        self.city = city

    def get_inf(self):
        return (f"Name: {self.name}\n"
                f"Surname: {self.surname}\n"
                f"Age: {self.age}\n"
                f"City: {self.city}")

    def get_name(self):
        return self.name

    def get_surname(self):
        return self.surname

    def get_age(self):
        return self.age

    def get_city(self):
        return self.city

    def __eq__(self, other):
        if self.age == other.age:
            return f"{self.name} and {other.name} are in the same age!"

    def __lt__(self, other):
        if self.age < other.age:
            return f"{self.name} is younger than {other.name}"

        elif self.age == other.age:
            return f"{self.name} and {other.name} are in the same age!"

        else:
            return f"{self.name} is older than {other.name}"

    def __repr__(self):
        return self.name

class Students:
    def __init__(self, name):
        self.name = name
        self.clas = []

    def __repr__(self):
        return self.name

    def __len__(self):
        amount = len(self.clas)
        return f"Students: {amount} amount"

    def __getitem__(self, index):
        return self.clas[index]

    def __setitem__(self, index, value):
        if isinstance(value, Student):
            self.clas[index] = value

    def add_student(self, *new):
        for n in new:
            if isinstance(n, Student):
                self.clas.append(n)
            else:
                print("Enter students!")

class Subjects(Student):

    def __init__(self, name, surname, age, city):
        super().__init__(name, surname, age, city)
        self.subjects = []


    def add_subjects(self, *subject):
        self.subjects.extend(subject)
        return f"{self.name}: {self.subjects}"

    def __add__(self, subjects):
        if subjects not in self.subjects:
            self.subjects.append(subjects)
        return self

    def __sub__(self, subjects):
        if subjects in self.subjects:
            self.subjects.remove(subjects)
        return self



students1 = Students("11B")
students2 = Students("11A")

student1 = Subjects("Islom", "Amanullaev", 17, "New York")
print(student1.get_inf())
student1 += "Art"
print(student1.add_subjects("Mathematics"))
print("_________________________")
student2 = Subjects("Liam", "O'Connor", 25, "Dushanbe")
print(student2.get_inf())
student2.add_subjects("Mathematics", "Art")
student2 -= "Art"
print(student2.name, student2.subjects)
students1.add_student(student1, student2)
print(f"In {students1} we have {students1.clas}")
print("                                 "
      "")
student3 = Subjects("Fernando", "Gonzalez", 22, "Tashkent")
print(student3.get_inf())
student3 += "Biology"
print(student3.add_subjects("Mathematics", "Computer Science"))
print("_________________________")
student4 = Subjects("Maria", "Rodriguez", 26, "Berlin")
print(student4.get_inf())
student4 +="CS"
print(student4.add_subjects("Artificial intelligence", "Data Science"))
students2.add_student(student3, student4)
print(f"In {students2} => {students2.clas}")
```
